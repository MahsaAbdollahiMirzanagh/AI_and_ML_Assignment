```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salary = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
#reading the first five lines
salary.head()
```

```python
#Target is salary
#Features is experience

y = salary["Salary"]
x = salary[["Experience Years"]]
```

```python
#deviding the data set into:
# 20% test
# 80% training


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8, random_state= 2529) 
```

```python
#printing the train and test Experience Years
print(x_train)
print(y_test)
```

```python
# creating and fitting the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
print(x_train)
print(y_test)

```

```python
# fitting the trained model 
model.fit(x_train,y_train)
```

```python
# finding the coefficient and intercept
print("coeff",model.coef_)
print(model.intercept_)
```

```python
#modeling prediction
y_predicted = model.predict(x_test)
```

```python
plt.plot(x_test, y_test,'o')
plt.plot(x_test,y_predicted)
plt.show()
```

```python
#calculating the errors between predicted and tested data
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_predicted)
```

```python
mean_absolute_percentage_error(y_test,y_predicted)
```

```python
mean_squared_error(y_test,y_predicted)
```
